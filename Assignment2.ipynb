###3)  Minmum and maximum age

###4) Number of retired people with cellular contact

````python
# No of retired people with cellular contact
df3 = data[['job','contact']]
df4 = df3[(df3['job']== 'retired') & (df3['contact']=='cellular')]
df4.head()
counts = len(df4.index)
print("No of retired people with cellular contact is %d" %counts)
````

###6) Number of people with a university degree 

````python
#No with university degree
df5= data['education'][data['education']=='university.degree']

size = len(df5.index)
print("No of people with uni degree is %d" %size)
````

###7) Pie chart representing different education

````python
#Pie chart of education qualifications
# Used data where values of education that are unknown were removed
# ed used from question 4 
countf=pd.DataFrame(data.groupby(by='education')['education'].count(),columns=['education'])


fig = px.pie(countf,names=countf.index, values= countf['education'],color_discrete_sequence=px.colors.sequential.Magma,
             title='Respresentation of educational qualifications',labels={'Qualification':'education'},
             hover_data= ['education'])
fig.show()
````

###7) Most frequent marital status

````python
# Most frequent marital status
mdf= pd.Series(data['marital'])
mcount =mdf.value_counts()
most =mcount.index[0]
print("Most frequent marital status is %s" %most)
````

###8) Least frequent marital status 

````python
#Least frequent martial status
mdf= pd.Series(data['marital'])
mcount =mdf.value_counts()
least =mcount.index[-1]

print("Least frequent marital status is %s" %least)
````
